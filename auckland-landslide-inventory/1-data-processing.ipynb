# Data Processing

```python
import pandas as pd
import geopandas as gpd
```

# Load data

```python
geo_filepath = 'data/landslide_data.gpkg'
gdf = gpd.read_file(geo_filepath, layer='landslides')
gdf.head()
```

```python
gdf.info()
```

## Columns Selection

```python
gdf = gdf[['GlobalID', 'geometry', 'latitude', 'longitude', 'region', 'materialtype', 'dateoccurence', 'dateconfidence', 'geometrymodifiedon','createdon']]

geo_selected_cols_filepath = 'data/landslide_data_selected_cols.gpkg'
gdf.to_file(geo_selected_cols_filepath, driver='GPKG', layer='landslides')
```

```python
# Convert columns to datetime
for col in ['dateoccurence', 'createdon', 'geometrymodifiedon', 'dateconfidence']:
    # gdf[col] = pd.to_datetime(gdf[col], errors='coerce')  # Convert invalid values to NaT
    print(gdf[col].apply(type).value_counts()) 
    
# print(gdf['dateconfidence'].apply(type).value_counts()) 
```

```python
# Convert columns to datetime
for col in ['dateoccurence', 'dateconfidence', 'createdon', 'geometrymodifiedon']:
    gdf[col] = pd.to_datetime(gdf[col], errors='coerce', utc=True)  # Convert invalid values to NaT
    gdf[col] = gdf[col].dt.tz_localize(None)  #
```

```python
gdf[gdf['dateconfidence'].notna() & (gdf['dateconfidence'] < gdf['dateoccurence'])][['dateconfidence','dateoccurence']]
```

```python
gdf['closeddate'] = gdf[['dateoccurence', 'createdon', 'dateconfidence', 'geometrymodifiedon']].min(axis=1)
```

```python
gdf.drop(columns=['dateoccurence', 'createdon', 'dateconfidence', 'geometrymodifiedon'], inplace=True)
```

```python
geo_cleanned_filepath = 'data/landslide_data_cleaned.gpkg'
gdf.to_file(geo_cleaned_filepath, driver='GPKG', layer='landslides')
```
